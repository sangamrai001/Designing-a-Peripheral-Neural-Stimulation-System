```python
import matplotlib.pyplot as plt
```

```python
# Values of scaling factor at different x coordinates
my_dict = {0.0: 0.0023, 0.02: 0.0023, -0.02: 0.0023, 0.004: 0.0023, -0.004: 0.0023, 0.04: 0.0013, -0.04 :0.0013 , 0.008 :0.0023,-0.008:0.0023,  0.012: 0.0023 , -0.012: 0.0023, 0.016: 0.0023, -0.016: 0.0023, 0.024: 0.0023, -0.024: 0.0023, 0.028: 0.0023, -0.028: 0.0023, 0.032: 0.0023, -0.032: 0.0023, 0.036: 0.0023, -0.036: 0.0023}

sorted_dict = dict(sorted(my_dict.items())) # sort the dictnary
k = []
val = []
for key, value in sorted_dict.items():
    k.append(key)
    val.append(value)

plt.plot(k, val,color = "red")
plt.xlabel('x coordinate')
plt.ylabel('Minimum Scaling factor')
plt.title('Plot of Scaling factor vs X-coordinate (z = 0.5 mm ) and (y = 4 mm)')
plt.show()
```

```python
# Values of scaling factor at different y coordinates
my_dict = { 0.001: 0.00093, -0.001: 0.00093, 0.01: 0.0368, -0.01: 0.0366, 0.002: 0.00085, -0.002: 0.00083, 0.003: 0.001, -0.003: 0.00094, 0.004: 0.0024, -0.004: 0.00245, 0.005: 0.0047, -0.005: 0.0047, 0.006: 0.0082, -0.006: 0.00815, 0.007: 0.013, -0.007: 0.0128, 0.008: 0.0195, -0.008: 0.019, 0.009: 0.0268, -0.009: 0.0267}
sorted_dict = dict(sorted(my_dict.items()))  # sort the dictnary
k = []
val = []
k1 = []
val2 = []
i = 0

# Loop is used here beacuse at y = 0 there is a discontinutiy in the plot 
for key, value in sorted_dict.items():
    if i <= 9:
        k.append(key)
        val.append(value)
    else:
        k1.append(key)
        val2.append(value)
    i += 1

plt.plot(k, val,color = "red")
plt.plot(k1, val2,color = "red")
plt.xlabel('y coordinate')
plt.ylabel('Minimum Scaling factor')
plt.title('Plot of Scaling factor V/s y coordinate where (x = 0 mm ) and (z = 0.5 mm)')
plt.show()
```
